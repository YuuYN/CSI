## Bài 1. Họ và Tên
**Input**: cho người dùng nhập vào họ tên.
**Output**: Họ và tên theo chuẩn tiếng Việt.  
**Ví dụ**:
- Input:

## Bài 2. Số Lớn Thứ Hai
**Input**: Cho người dùng nhập vào Số phần tử (n) họ mong muốn cho 1 danh sách, yêu cầu *n > 0* và *n* số nguyên, từ n phần tử đó hãy random ra danh sách có các phần tử trong phạm vi từ 1 đến 20. Tìm ra số lớn thứ hai.
**Output**: Số lớn thứ hai trong các số đã nhập. In ra "Not found" nếu không tìm thấy.  
**Ví dụ**:  
Ví dụ 1:
- Input: n = 4, arr = [5, 4, 1, 12]
- Output: result = 5
- Giải thích: Số lớn nhất là 12, số lớn thứ hai là 5

Ví dụ 2:
- Input: n = 4, arr = [5, 4, 1, 12, 5, 12]
- Output: result = 5
- Giải thích: Có hai số cùng lớn nhất là 12, hai số cùng lớn thứ hai là 5

```python
import random

n = int(input('n ='))
print('n = ',n)
arr =[]
for i in range(n):
    random_number = random.randint(1, 20)
    arr.append(random_number)
print('arr = ',arr)


max_number1 = arr[0]
soluongnumber1 = 0
for number in arr:
    if number > max_number1:
        max_number1 = number

for number in arr:
    if number == max_number1:
        arr.remove(max_number1)
        soluongnumber1 = soluongnumber1 + 1

    
print('arr = ', arr)

max_number2 = arr[0]
soluongnumber2 = 0
for number in arr:
    if number > max_number2:
        max_number2 = number

for number in arr:
    if number == max_number2:
        soluongnumber2 = soluongnumber2 + 1

if soluongnumber1 > 1 or soluongnumber2 > 1:
    print('có ',soluongnumber1, ' so lon nhat là ',max_number1,
          ', có ',soluongnumber2, ' so lon thu 2 là ',max_number2)
else :
    print('so lon nhat là ',max_number1,', so lon thu 2 là ',max_number2)



```
